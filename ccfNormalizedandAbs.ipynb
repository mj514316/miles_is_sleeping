```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def make_event(n,start,dur):

    return [1 if (x > start and x < start + dur+1) else 0 for x in range(L) ]
    #[unicode(x.strip()) if x is not None else '' for x in row]
def normcorr(df,var1,var2):
    ccfcorr = np.correlate(df[var1]-np.mean(df[var1]),df[var2]-np.mean(df[var2]), 'Full')/df.shape[0]/(np.std(df[var1]) * np.std(df[var2]))
    return ccfcorr
#make_event(L,20,5)
```

```python
L = 50
df = pd.DataFrame(np.random.normal(size = (L, 2)), columns=list('AB'))

df['increase'] = [x for x in range(L)]

df['A_inc'] = df['A']+df['increase']*.1
df['B_inc'] = df['B']+df['increase']*.1
df['B_dec'] = df['B']-df['increase']*.1


df['A_e_20_5'] = df['A']+np.array(make_event(L,20,5))*10
df['Bdec_e_30_5'] = df['B']-np.array(make_event(L,30,5))*10
df['Binc_e_30_5'] = df['B']+np.array(make_event(L,30,5))*10

df
```

```python
df[['A','B']].plot()
```

```python
normcorr(df,'A','B').round(1)
```

```python
df[['A_inc','B_dec']].plot()
```

```python
nc = normcorr(df,'A_inc','B_dec')
print(len(nc))
print(f" max coef = {np.max(nc).round(2)}, min coef = {np.min(nc).round(2)}, argmax: {np.argmax(nc)-L+1}, argmax(abs): {np.argmax(np.abs(nc))-L+1}")
plt.scatter([x for x in range(L*2-1)],nc)
```
